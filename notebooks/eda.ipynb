# Univariate Analysis

```python
from visualization import histogram_and_box_plot 
from cleaning import calculate_outliers

def plot_visualization_and_stats(column_name: str, df: pd.DataFrame) ->None:
    histogram_and_box_plot(column_name,df)
    print(f"Statistics for {column_name}")
    print(f"Skewness: {df[column_name].skew():.4f}")
    lower_bound, upper_bound = calculate_outliers(df,column_name)
    print("Outliers")
    print(f" less than lower bound: {df[df[column_name] < lower_bound][column_name]}")
    print(f" higher than upper bound: {list(df[(df[column_name]>upper_bound)][column_name])}")
    print(f"standard deviation: {df[column_name].std()}")
    print(f"Mean :{df[column_name].mean()} ")
    print(f"CV :{(df[column_name].std()/df[column_name].mean()):.4f} ")


for col in df.columns:
    plot_visualization_and_stats(col, df)
```

#Observation:

* Most features exhibit heavy right skewness, indicating the presence of significant outliers far from the mean.

* These outliers are not noise but rather reflect rare, high-impact events in certain ZIP codes. For example, out of 2,118 ZIP codes, 2,093 had between 0â€“19 claims. However, around 15 ZIP codes recorded more than 63 claims, with some of the most severely affected areas reporting over 200 claims.

* These high-impact events could be key drivers of next year's premium rates in the affected ZIP codes.

* Rather than removing them entirely, we can apply capping (also known as winsorization) to retain meaningful outliers while limiting the influence of extreme values that may not add predictive value.
